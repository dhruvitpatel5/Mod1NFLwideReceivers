```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Loading Data
file_path = 'FantasyPros_Fantasy_Football_Statistics_WR.csv' 
wr_df = pd.read_csv(file_path) 
wr_df.info()

```

```python
# Clean data 
wr_df.dropna(inplace=True)
wr_df['YDS'] = wr_df['YDS'].str.replace(',','').astype(float)
wr_df.head()

```

```python
# Graph data (YDS and TD)
plt.scatter(wr_df['YDS'],wr_df['TD'], c='green',marker = '.')
plt.xlabel('Receiving Yards')
plt.ylabel('Touchdowns')
plt.title('Receiving Yards (of WR) vs. Touchdowns in 2023')
plt.grid(linestyle=':')
plt.show()

```
